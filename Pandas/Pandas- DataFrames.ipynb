# 1.1 With creatation of python dictionary -1

```python
import pandas as pd

'Using of Pythons dicctionary'
weather_data = {
    'day': ['1/1/2017','1/2/2017','1/3/2017','1/4/2017','1/5/2017','1/6/2017'], # Row values
    'temperature': [32,35,28,24,32,31],# Row values# Row values
    'windspeed': [6,7,2,7,4,2],# Row values
    'event': ['Rain', 'Sunny', 'Snow','Snow','Rain', 'Sunny']# Row values
}
'Decleration of DataFrame'
df=pd.DataFrame(weather_data) 
df
```

# 1.1 With creatation of python dictionary -2

```python
weather_data = [
    {'day': '1/1/2017', 'temperature': 32, 'windspeed': 6, 'event': 'Rain'}, # Column values
    {'day': '1/2/2017', 'temperature': 35, 'windspeed': 7, 'event': 'Sunny'}, # Column values
    {'day': '1/3/2017', 'temperature': 28, 'windspeed': 2, 'event': 'Snow'}, # Column values
    
]
df = pd.DataFrame(data=weather_data, columns=['day','temperature','windspeed','event']) #Define of header
df
```

#### 1.1.1 -  Highlighten of maximum Values

```python
'Defination of function'
def highlight_max(s):
    '''
    highlight the maximum in a Series yellow.
    '''
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max] 
```

```python
df.style.apply(highlight_max,subset='windspeed')
```

### 1.2  With creatation of *tuple list*

```python
weather_data = [
    ('1/1/2017',32,6,'Rain'),  # Each element in the list is touple
    ('1/2/2017',35,7,'Sunny'),
    ('1/3/2017',28,2,'Snow')
]

'Creation of header'
df = pd.DataFrame(data=weather_data, columns=['day','temperature','windspeed','event'])
df
```

### 1.4.2 Write two dataframes to two separate sheets in excel

```python
'DataFrame-1'
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})


'DataFrame-2'
df_weather =  pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})


with pd.ExcelWriter('stocks_weather.xlsx') as writer: #With with function
    df_stocks.to_excel(writer, sheet_name="stocks")
    df_weather.to_excel(writer, sheet_name="weather")
```

## 2 - Operation in Pandas

### 2.1 - Maximum number in a column

```python
'Whenn we wanna know a maximum number in a columns of DataFrame'
df['WindSpeedMPH'].max()
```

### 2.2 - Avarage of Numbers in a *column*

```python
' Avarage of Numbers in a column'
df['WindSpeedMPH'].mean()
```

### 2.3 - *Standard deviation* of numbers in a columns

```python
'Calculation of Standard deviation of numbers'
df['WindSpeedMPH'].std()
```

### 2.4 - Generate fo *descriptive statistics* of DataFrame

```python
'Generate descriptive statistics'
df.describe() # This is very useful for quickly printing of statistocs of our data
```

### 2.5. Flitering of Data in Columns

```python
'Filtering of columns regards of Temptreture less or equal 20 (<=20)'
df[df.Temperature <= 20]
```

## 2.5 -  Filtering of Data in *individual Column*

```python
'Filtering of columns <<Tepmreture>> regards of Temptreture less or equal 20 (<=20)'
df['Temperature'][df.Temperature<=20]
```

```python
'Filtering of columns <<Tepmreture>> regards of maximum Temptreture'
df[df.Temperature == df['Temperature'].max()] # This is usefull if we have space in the Name of our headers
```

```python
'If wee wanna see in wich date and which Events was the Tepmreture maximum'
df[['EST','Events']][df.Temperature == df['Temperature'].max()]
```
